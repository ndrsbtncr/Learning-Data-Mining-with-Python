**Extraer las categorias de los títulos de las noticias**

Categorias: categories: Macroeconomia, Sostenibilidad, Innovación, Regulaciones, Alianzas, Reputación

```python
import pandas as pd
```

```python
news=pd.read_csv('noticias.csv',encoding='utf-8')
```

```python
news.head()
```

```python
news_titulo=news[['news_title']]
news_titulo.head()
```

```python
filtro_macroeconomia = news_titulo[news_titulo['news_title'].str.contains("macroeconomi", case=False)]
filtro_macroeconomia.head()
```

```python
filtro_sostenibilidad = news_titulo[news_titulo['news_title'].str.contains("sostenibilidad", case=False)]
filtro_sostenibilidad.head()
```

```python
filtro_innovacion = news_titulo[news_titulo['news_title'].str.contains("innovacion", case=False)]
filtro_innovacion.head()
```

```python
filtro_regulaciones = news_titulo[news_titulo['news_title'].str.contains("regulacion", case=False)]
filtro_regulaciones.head()
```

```python
filtro_alianzas = news_titulo[news_titulo['news_title'].str.contains("alianza", case=False)]
filtro_alianzas.head()
```

```python
filtro_reputacion = news_titulo[news_titulo['news_title'].str.contains("reputacion", case=False)]
filtro_reputacion.head()
```
